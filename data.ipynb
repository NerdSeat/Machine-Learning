```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('C:/Projects/Machine-Learning/data/cost_revenue_clean.csv')
```

```python
data.describe(include='all')
```

```python
X = DataFrame(data['Production_Budget_usd'])
y = DataFrame(data['Worldwide_Gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,c='blue',alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

```python
regression.coef_ #theta_1 or slope
```

```python
regression.intercept_ #intercept
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,c='blue',alpha=0.3)
plt.plot(X,regression.predict(X),color='red',linewidth=3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
r_squared = regression.score(X,y)
r_squared
```
