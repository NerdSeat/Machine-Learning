```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('./data/lsd_math_score_data.csv')
```

```python
LSD = data[['LSD_ppm']]
scores = data[['Avg_Math_Test_Score']]
time=data[['Time_Delay_in_Minutes']]
```

```python
plt.title("Tissue concentration over Time",fontsize=14)
plt.xlabel('Time in minutes',fontsize=14)
plt.ylabel('Tissue LSD ppm',fontsize=14)
plt.text(x=0,y=-0.5,s='Wagner et al.(1968)',fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.ylim(1,7)
plt.xlim(0,500)

plt.style.use('classic')
plt.plot(time,LSD,color='r')
plt.show()
```

```python
regr = LinearRegression()
regr.fit(LSD,scores)
print('Theta1: ',regr.coef_[0][0])
print('Intercept: ',regr.intercept_[0])
print('R-squared: ',regr.score(LSD,scores))
predicted_score = regr.predict(LSD)
```

```python
%matplotlib inline
plt.title("Arithmetic vd LSD-25",fontsize=20)
plt.ylabel("Perfomance",fontsize=14)
plt.xlabel("Tissue LSD ppm",fontsize=14)
plt.style.use('fivethirtyeight')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(1,7)
plt.ylim(20,85)
plt.scatter(LSD,scores,color='blue',s=100,alpha=0.7)
plt.plot(LSD,predicted_score,color='green',linewidth=3)
plt.show()
```
