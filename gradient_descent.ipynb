# Notebook imports packages

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Example 1 - A simple cost notation
### $f(x) = x^2 + x+1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start=-3,stop=3,num=100)
```

```python
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('x_1',fontsize=16)
plt.ylabel('f(x_1)',fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

## Slope and Derivatives

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize =(15,5))

# Chart 1: Cost Function
plt.subplot(1,2,1)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.title('Cost Function')
plt.xlabel('x_1',fontsize=16)
plt.ylabel('f(x_1)',fontsize=16)

plt.plot(x_1,f(x_1),color ='blue',linewidth=3)

# Chart 2: Derivative of the cost function
plt.subplot(1,2,2)
plt.title('Derivative of the Cost Function')

plt.xlabel('x_1',fontsize=16)
plt.ylabel('f \'(x_1)',fontsize=16)
plt.plot(x_1,df(x_1),color='skyblue',linewidth=5)
plt.xlim(-2,3)
plt.ylim(-2,6)
plt.grid()
plt.show()
```
