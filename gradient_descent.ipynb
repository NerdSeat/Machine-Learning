# Notebook imports packages

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Example 1 - A simple cost notation
### $f(x) = x^2 + x+1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start=-3,stop=3,num=100)
```

```python
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('x_1',fontsize=16)
plt.ylabel('f(x_1)',fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

## Slope and Derivatives

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize =(15,5))

# Chart 1: Cost Function
plt.subplot(1,2,1)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.title('Cost Function')
plt.xlabel('x_1',fontsize=16)
plt.ylabel('f(x_1)',fontsize=16)

plt.plot(x_1,f(x_1),color ='blue',linewidth=3)

# Chart 2: Derivative of the cost function
plt.subplot(1,2,2)
plt.title('Derivative of the Cost Function')

plt.xlabel('x_1',fontsize=16)
plt.ylabel('f \'(x_1)',fontsize=16)
plt.plot(x_1,df(x_1),color='skyblue',linewidth=5)
plt.xlim(-2,3)
plt.ylim(-2,6)
plt.grid()
plt.show()
```

# Gradient Descent

```python
new_x = -3
previous = 0
step_multiplier = 0.1
precision = 0.0001

x_list = []
slope_list=[]
for i in range(500):
    previous=new_x
    gradient=df(previous)
    new_x = previous - step_multiplier*gradient
    step_size = abs(new_x-previous)
    #print(i,step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))
    if(step_size<=precision):
        break
print('local minimum occurs at',new_x)
print('f \'(x) at this point is ',df(new_x))
print('f(x) at this point is ',f(new_x))
```

```python
# Super impose the gradient descent

plt.figure(figsize =(25,6))

# Chart 1: Cost Function
plt.subplot(1,3,1)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.title('Cost Function')
plt.xlabel('X',fontsize=16)
plt.ylabel('f(X)',fontsize=16)

plt.plot(x_1,f(x_1),color ='blue',linewidth=3)
# Convert the list into an array before parsing it to the function
values = np.array(x_list)
plt.scatter(x_list,f(values),color='green',s=100,alpha=0.7)

# Chart 2: Derivative of the cost function
plt.subplot(1,3,2)
plt.title('Derivative of the Cost Function')

plt.xlabel('X',fontsize=16)
plt.ylabel('f \'(X)',fontsize=16)
plt.plot(x_1,df(x_1),color='black',linewidth=4,alpha=0.6)
plt.xlim(-2,3)
plt.ylim(-2,6)

val = np.array(slope_list)
plt.scatter(val,df(val),color='red',s=100,alpha=0.5)
plt.grid()

# Chart 3: Derivative of the cost function close up
plt.subplot(1,3,3)
plt.title('Derivative of the Cost Function (Close up)')

plt.xlabel('X',fontsize=16)
plt.ylabel('slope',fontsize=16)
plt.plot(x_1,df(x_1),color='skyblue',linewidth=6,alpha=0.6)
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.scatter(x_list,slope_list,color='red',s=300,alpha=0.5)

plt.show()
```
